```python
try:
  sayi = int(input("Bir sayı gir!!"))
  print(sayi)
except ValueError:
  print("Hata: Girilen veri sayı değildir!!")
else:
  print("Giriş işlemi başarılı")
finally:
  print("Program sonlandırıldı.")
```

```python
try:
  s = 5 / 0
  degisken = int(s)
except ZeroDivisionError:
  print("Dönüşüm Hatası")
else:
  print("işlem tamamlandı", degisken)
finally:
  print("program bitti")
```

```python
while True:
  girdi = input("Bir sayı gir(Çıkmak için 'q' bas)")
  if girdi.lower() == 'q':
    print("Programdan Çıkılıyor")
    break
  try:
    sayi = float(girdi)
  except ValueError:
    print("Hata: Girilen veri sayı değildir!!")
    continue

  if sayi == 0:
    print("Sayı sıfırdır")
  elif sayi>0:
    print("Sayı pozitif")
  else:
    print("Sayı negatif")

  break
```
